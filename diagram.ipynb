```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare

# Thiết lập định dạng cho biểu đồ
plt.rcParams.update({
    "figure.dpi": 150,
    "font.size": 11,
})

# =============================================================================
# DATA DEFINITIONS (v06 & v07)
# =============================================================================
xgbo_v06 = np.array([[6.11789608e-03, 4.06582343e-05, 2.56373211e-03, 7.86757923e-01], [5.88863124e-03, 3.74558124e-05, 2.55512428e-03, 8.02482121e-01], [5.92400186e-03, 3.68940336e-05, 2.55256354e-03, 7.98115596e-01], [5.89557368e-03, 3.64830889e-05, 2.55727595e-03, 7.99412479e-01], [5.92631658e-03, 3.75172084e-05, 2.55764964e-03, 7.99805341e-01], [5.91111282e-03, 3.80614049e-05, 2.55817868e-03, 8.00373843e-01], [5.86726996e-03, 3.70165200e-05, 2.55616933e-03, 8.02231135e-01], [5.91635367e-03, 3.73707455e-05, 2.55911517e-03, 7.97660193e-01], [5.84656402e-03, 3.59101852e-05, 2.55682002e-03, 8.03515100e-01], [5.80888648e-03, 3.61341777e-05, 2.55217503e-03, 8.07345152e-01], [5.87366877e-03, 3.69938398e-05, 2.55543562e-03, 8.03038986e-01], [5.77542418e-03, 3.71568345e-05, 2.55466388e-03, 8.12234040e-01], [5.89270328e-03, 3.67195330e-05, 2.55587298e-03, 8.01010009e-01], [5.90221148e-03, 3.66601941e-05, 2.55543144e-03, 7.97509900e-01], [5.89123275e-03, 3.66349749e-05, 2.55746794e-03, 7.99724763e-01], [5.91905993e-03, 3.75155448e-05, 2.55523925e-03, 7.97921534e-01], [6.00221848e-03, 3.92331980e-05, 2.55392593e-03, 7.95417627e-01], [5.90946824e-03, 3.65642037e-05, 2.55461294e-03, 7.97741972e-01], [5.87934105e-03, 3.67103978e-05, 2.55772166e-03, 8.00505050e-01], [5.90438694e-03, 3.75736098e-05, 2.56553900e-03, 8.01143323e-01]])
cat_v06 = np.array([[6.45286910e-03, 4.52280068e-05, 3.03967387e-03, 7.64331431e-01], [6.43787348e-03, 4.43345621e-05, 3.03203964e-03, 7.64111748e-01], [6.40475616e-03, 4.38153559e-05, 3.03339898e-03, 7.65777314e-01], [6.34688363e-03, 4.22435694e-05, 3.03077474e-03, 7.68191525e-01], [6.43088545e-03, 4.48880461e-05, 3.03432027e-03, 7.65546140e-01], [6.35114289e-03, 4.32162556e-05, 3.03267724e-03, 7.69171615e-01], [6.39479427e-03, 4.39338856e-05, 3.03404659e-03, 7.66298791e-01], [6.40596821e-03, 4.40362130e-05, 3.04133578e-03, 7.66325011e-01], [6.39291777e-03, 4.31902134e-05, 3.04925923e-03, 7.65973332e-01], [6.39468843e-03, 4.39008523e-05, 3.04342673e-03, 7.67081339e-01], [6.32071373e-03, 4.27661403e-05, 3.02884556e-03, 7.72308291e-01], [6.35058247e-03, 4.51346530e-05, 3.02931030e-03, 7.73976303e-01], [6.33906514e-03, 4.24398927e-05, 3.03197790e-03, 7.70189114e-01], [6.38695935e-03, 4.31004253e-05, 3.03220814e-03, 7.65584822e-01], [6.32034503e-03, 4.23885139e-05, 3.02963869e-03, 7.71203278e-01], [6.45203979e-03, 4.44626617e-05, 3.06391150e-03, 7.61627959e-01], [6.45694182e-03, 4.56895707e-05, 3.03417966e-03, 7.64358666e-01], [6.32032465e-03, 4.22928371e-05, 3.03147005e-03, 7.70909151e-01], [6.36823240e-03, 4.31094821e-05, 3.02557110e-03, 7.67740046e-01], [6.39764287e-03, 4.37741262e-05, 3.04333491e-03, 7.66813502e-01]])
random_v06 = np.array([[8.34206421e-03, 7.25223585e-05, 5.02711467e-03, 5.98990527e-01], [8.32238242e-03, 7.25245318e-05, 5.02303612e-03, 6.01627888e-01], [8.36401749e-03, 7.33738015e-05, 5.02786339e-03, 5.99166798e-01], [8.32286382e-03, 7.18008249e-05, 5.02588226e-03, 6.00810758e-01], [8.49035163e-03, 7.72045559e-05, 5.02969777e-03, 5.90832650e-01], [8.32157462e-03, 7.15044999e-05, 5.02370079e-03, 5.99154243e-01], [8.35791644e-03, 7.31838188e-05, 5.02925595e-03, 5.99258634e-01], [8.34916268e-03, 7.37959716e-05, 5.02168676e-03, 6.02518724e-01], [8.47017259e-03, 7.43966392e-05, 5.02401745e-03, 5.84954446e-01], [8.39420633e-03, 7.31381191e-05, 5.03385853e-03, 5.93886629e-01], [8.33698099e-03, 7.20013252e-05, 5.02735974e-03, 5.98467104e-01], [8.41673784e-03, 7.41678263e-05, 5.03369412e-03, 5.93653864e-01], [8.35205054e-03, 7.22092962e-05, 5.03067993e-03, 5.97204384e-01], [8.50834790e-03, 7.73275523e-05, 5.05556292e-03, 5.91855065e-01], [8.35074888e-03, 7.25923624e-05, 5.02581160e-03, 5.98271966e-01], [8.37228466e-03, 7.36162399e-05, 5.03236251e-03, 5.97932729e-01], [8.48712048e-03, 7.48465336e-05, 5.03741935e-03, 5.84346241e-01], [8.32489351e-03, 7.23220401e-05, 5.03113637e-03, 6.01088295e-01], [8.33648106e-03, 7.19717591e-05, 5.02191550e-03, 5.98172415e-01], [8.40151464e-03, 7.33096456e-05, 5.02406761e-03, 5.92450713e-01]])
lgbm_v06 = np.array([[6.73069920e-03, 4.83829182e-05, 3.43204023e-03, 7.42595008e-01], [6.71379672e-03, 4.79891370e-05, 3.43119784e-03, 7.42975253e-01], [6.75455706e-03, 4.83423374e-05, 3.43648130e-03, 7.39551722e-01], [6.70079156e-03, 4.71435424e-05, 3.43306879e-03, 7.42256227e-01], [6.77707722e-03, 4.89408245e-05, 3.43615393e-03, 7.38504410e-01], [6.70687759e-03, 4.77045410e-05, 3.43309909e-03, 7.42724766e-01], [6.70689779e-03, 4.76826690e-05, 3.43594064e-03, 7.42666412e-01], [6.71058970e-03, 4.81203336e-05, 3.43164689e-03, 7.43530903e-01], [6.69065124e-03, 4.70813731e-05, 3.43483334e-03, 7.43756397e-01], [6.74746772e-03, 4.88350055e-05, 3.44410884e-03, 7.41347662e-01], [6.69389419e-03, 4.76777109e-05, 3.42911037e-03, 7.44629245e-01], [6.74094949e-03, 4.93780290e-05, 3.42814233e-03, 7.43536517e-01], [6.70825141e-03, 4.73933908e-05, 3.43027615e-03, 7.42260342e-01], [6.74826474e-03, 4.79788481e-05, 3.43055359e-03, 7.38577161e-01], [6.72773502e-03, 4.78639220e-05, 3.43681938e-03, 7.41126712e-01], [6.76809884e-03, 4.86236248e-05, 3.42771768e-03, 7.37937334e-01], [6.77577301e-03, 4.98920556e-05, 3.43557596e-03, 7.40334073e-01], [6.74335384e-03, 4.78076223e-05, 3.42996373e-03, 7.39332003e-01], [6.72244833e-03, 4.76582838e-05, 3.44107557e-03, 7.41248882e-01], [6.68586882e-03, 4.75217523e-05, 3.43322282e-03, 7.45066244e-01]])
gra_v06 = np.array([[5.57043611e-03, 3.31932322e-05, 2.76549074e-03, 8.22976564e-01], [5.62744475e-03, 3.33952387e-05, 2.77093526e-03, 8.17260571e-01], [5.63492312e-03, 3.41907744e-05, 2.76725575e-03, 8.18930920e-01], [5.63906297e-03, 3.46373391e-05, 2.76061122e-03, 8.19721353e-01], [5.67601345e-03, 3.47698849e-05, 2.78021368e-03, 8.18061991e-01], [5.70156001e-03, 3.49688728e-05, 2.77062992e-03, 8.16155685e-01], [5.61175922e-03, 3.33063437e-05, 2.76347967e-03, 8.19331240e-01], [5.62803618e-03, 3.37573998e-05, 2.76614368e-03, 8.19503471e-01], [5.59753266e-03, 3.34689431e-05, 2.77062765e-03, 8.20603479e-01], [5.75335701e-03, 3.73356399e-05, 2.76008541e-03, 8.14711048e-01], [5.59879968e-03, 3.38111730e-05, 2.76568090e-03, 8.21636158e-01], [5.59794885e-03, 3.33529650e-05, 2.76640198e-03, 8.20506768e-01], [5.74008943e-03, 3.51679232e-05, 2.76438444e-03, 8.12150818e-01], [5.65731474e-03, 3.40965531e-05, 2.76253246e-03, 8.17679701e-01], [5.64581298e-03, 3.42047723e-05, 2.78210428e-03, 8.20234531e-01], [5.78220315e-03, 3.58160119e-05, 2.77474694e-03, 8.09429312e-01], [5.61924982e-03, 3.49245559e-05, 2.76503587e-03, 8.22487884e-01], [5.66562852e-03, 3.44775638e-05, 2.77295840e-03, 8.17455751e-01], [5.60373972e-03, 3.37086223e-05, 2.76235104e-03, 8.21378357e-01], [5.56777633e-03, 3.25900199e-05, 2.76105004e-03, 8.21960670e-01]])
procat_v06 = np.array([[5.56544562e-03, 3.27274952e-05, 2.30997385e-03, 8.19712366e-01], [5.52037382e-03, 3.27273119e-05, 2.31173908e-03, 8.23296848e-01], [5.54803360e-03, 3.32828669e-05, 2.31215035e-03, 8.22403623e-01], [5.54755515e-03, 3.27142746e-05, 2.31080634e-03, 8.21844799e-01], [5.47010821e-03, 3.17685977e-05, 2.30954576e-03, 8.25304682e-01], [5.51920793e-03, 3.29315312e-05, 2.31394735e-03, 8.23946933e-01], [5.54697949e-03, 3.27515918e-05, 2.31456370e-03, 8.21481121e-01], [5.51259606e-03, 3.30798527e-05, 2.31296797e-03, 8.25433877e-01], [5.46410413e-03, 3.12982503e-05, 2.30891240e-03, 8.25471207e-01], [5.54267725e-03, 3.31539464e-05, 2.31263629e-03, 8.22054481e-01], [5.49868431e-03, 3.18715544e-05, 2.31004681e-03, 8.22896852e-01], [5.44005653e-03, 3.06853923e-05, 2.31148905e-03, 8.25011561e-01], [5.54013709e-03, 3.43623165e-05, 2.31203192e-03, 8.25103366e-01], [5.48971271e-03, 3.14336531e-05, 2.31196184e-03, 8.23060655e-01], [5.43865709e-03, 3.12445321e-05, 2.31139825e-03, 8.26539951e-01], [5.52925688e-03, 3.23837037e-05, 2.31228160e-03, 8.19647421e-01], [5.57551607e-03, 3.36499660e-05, 2.31281572e-03, 8.21201640e-01], [5.51924691e-03, 3.23628091e-05, 2.30803152e-03, 8.22546567e-01], [5.47248777e-03, 3.25988316e-05, 2.31214590e-03, 8.27993026e-01], [5.54647358e-03, 3.33396753e-05, 2.31124799e-03, 8.22297229e-01]])

cat_v07 = np.array([[4.81808869e-03, 3.84619727e-05, 1.45477382e-03, 7.89598633e-01], [4.72264349e-03, 3.68551939e-05, 1.39339574e-03, 7.97795615e-01], [4.88110633e-03, 3.80321901e-05, 1.56848894e-03, 7.81218294e-01], [4.65868807e-03, 3.37589020e-05, 1.48985133e-03, 8.03297644e-01], [5.05318364e-03, 4.53446089e-05, 1.45615039e-03, 7.66392507e-01], [4.85734793e-03, 4.05909482e-05, 1.45515417e-03, 7.84309317e-01], [4.65438240e-03, 3.55096868e-05, 1.39278499e-03, 8.04544766e-01], [4.92170384e-03, 3.88392747e-05, 1.50589849e-03, 7.77469547e-01], [4.88471048e-03, 3.77138599e-05, 1.52754158e-03, 7.81281129e-01], [2.03269111e-02, 3.14135468e-03, 2.83577701e-03, 8.07446802e-01], [4.70428305e-03, 3.65962414e-05, 1.39608678e-03, 8.00288942e-01], [4.83989325e-03, 3.77980901e-05, 1.49881247e-03, 7.85893838e-01], [5.03354040e-03, 4.58811397e-05, 1.46675063e-03, 7.67632349e-01], [5.02836275e-03, 3.98647585e-05, 1.56177859e-03, 7.65942317e-01], [4.85726138e-03, 3.61626613e-05, 1.51525242e-03, 7.83518726e-01], [4.72045813e-03, 3.69627900e-05, 1.39703323e-03, 7.98804818e-01], [4.89719507e-03, 3.66808848e-05, 1.59746999e-03, 7.78505754e-01], [4.91972304e-03, 4.02576425e-05, 1.49848233e-03, 7.79735098e-01], [5.07513844e-03, 4.48738054e-05, 1.53399980e-03, 7.63645435e-01], [4.90824525e-03, 3.94486942e-05, 1.52900195e-03, 7.76289725e-01]])
grad_v07 = np.array([[5.16447810e-03, 4.69037517e-05, 1.26667935e-03, 7.51942389e-01], [5.50239671e-03, 4.90236021e-05, 1.27060491e-03, 6.91791939e-01], [5.68845489e-03, 5.24938534e-05, 1.26400009e-03, 6.46203783e-01], [5.57446657e-03, 5.06405623e-05, 1.26344806e-03, 6.87237252e-01], [5.67606358e-03, 5.10679576e-05, 1.27428771e-03, 6.73980184e-01], [6.81865823e-03, 8.31872726e-05, 1.29527484e-03, 2.24428161e-01], [5.97299563e-03, 5.38894346e-05, 1.26187751e-03, 6.29043902e-01], [6.14075554e-03, 5.86811753e-05, 1.27740302e-03, 5.65513073e-01], [5.97990721e-03, 5.40892167e-05, 1.26848068e-03, 6.27589754e-01], [5.14380550e-03, 4.58089981e-05, 1.26103393e-03, 7.50273448e-01], [5.40588249e-03, 4.76130320e-05, 1.25955070e-03, 7.18848837e-01], [5.57560616e-03, 4.97064838e-05, 1.27257040e-03, 6.96715411e-01], [4.90671333e-03, 4.27090890e-05, 1.25461974e-03, 7.80325222e-01], [6.09800671e-03, 5.64132789e-05, 1.27965988e-03, 6.02784269e-01], [5.61203931e-03, 5.09224185e-05, 1.26687601e-03, 6.80802945e-01], [4.86498894e-03, 4.32638468e-05, 1.25009969e-03, 7.82114894e-01], [6.66552518e-03, 8.56380379e-05, 1.30534533e-03, 2.65635690e-01], [6.07248340e-03, 6.01823551e-05, 1.28074324e-03, 5.35603462e-01], [2.07097857e-02, 3.10259374e-03, 2.53856573e-03, 7.28923091e-01], [5.47587995e-03, 4.98468073e-05, 1.26308188e-03, 6.92948846e-01]]) 
light_v07 = np.array([[4.53940578e-03, 2.98184268e-05, 1.47380292e-03, 8.15751382e-01], [4.66272432e-03, 3.11663705e-05, 1.46198562e-03, 8.04410395e-01], [5.39150923e-03, 3.95392925e-05, 1.49107378e-03, 6.46108286e-01], [4.76474041e-03, 2.96802612e-05, 1.49528989e-03, 7.90705836e-01], [5.24247602e-03, 4.16398964e-05, 1.48712113e-03, 7.48774665e-01], [5.01892989e-03, 3.73366722e-05, 1.48644060e-03, 7.64675733e-01], [4.89747385e-03, 3.23218652e-05, 1.47870615e-03, 7.75853702e-01], [5.52680753e-03, 4.30310789e-05, 1.48959983e-03, 6.35604198e-01], [4.66490339e-03, 2.71023970e-05, 1.48217366e-03, 7.95882804e-01], [2.06340372e-02, 3.12865040e-03, 2.81303590e-03, 7.75849901e-01], [4.80689751e-03, 3.12563526e-05, 1.47950768e-03, 7.89074397e-01], [5.01256445e-03, 3.48461758e-05, 1.48868926e-03, 7.69457848e-01], [5.30698456e-03, 4.78593714e-05, 1.49566168e-03, 7.40985338e-01], [6.02011726e-03, 5.28179723e-05, 1.49830036e-03, 5.87588506e-01], [5.76049031e-03, 4.67641239e-05, 1.49824157e-03, 5.85601788e-01], [4.91988590e-03, 3.40018981e-05, 1.50114646e-03, 7.80295947e-01], [5.33695953e-03, 3.77379004e-05, 1.49157632e-03, 6.52249038e-01], [4.73855861e-03, 3.13940250e-05, 1.48108362e-03, 7.95357773e-01], [5.98519331e-03, 5.23108392e-05, 1.51141049e-03, 5.87674628e-01], [5.71530520e-03, 4.51738883e-05, 1.50394492e-03, 6.16728415e-01]])
ran_v07 = np.array([[6.68726469e-03, 6.32826247e-05, 1.89863952e-03, 5.44203536e-01], [6.20478161e-03, 5.43199689e-05, 1.88708391e-03, 6.19097449e-01], [6.77024906e-03, 6.19321293e-05, 1.89335132e-03, 5.35809508e-01], [6.63170873e-03, 5.94297207e-05, 1.89468293e-03, 5.21781782e-01], [6.46703456e-03, 5.85469194e-05, 1.89127578e-03, 5.82202213e-01], [6.71586562e-03, 6.18027897e-05, 1.89149200e-03, 5.46881686e-01], [6.51051760e-03, 5.87679293e-05, 1.88961279e-03, 5.70435557e-01], [2.18321068e-02, 3.21278791e-03, 3.61316948e-03, 6.40758739e-01], [6.76060943e-03, 6.25838049e-05, 1.89344821e-03, 5.31468737e-01], [6.14158553e-03, 5.38065129e-05, 1.87327495e-03, 6.20216636e-01], [6.96433726e-03, 7.07243175e-05, 1.90368135e-03, 4.14118690e-01], [6.23578631e-03, 5.50671172e-05, 1.88647032e-03, 6.20918139e-01], [6.10561541e-03, 5.52233261e-05, 1.88911305e-03, 6.47061044e-01], [6.52854151e-03, 5.73689532e-05, 1.88869775e-03, 5.69337819e-01], [6.10209987e-03, 5.26460579e-05, 1.88467675e-03, 6.41031556e-01], [6.17485455e-03, 5.34143844e-05, 1.89058707e-03, 6.16772071e-01], [6.02226215e-03, 5.38838960e-05, 1.88238900e-03, 6.64287621e-01], [6.82932625e-03, 6.24986867e-05, 1.89429817e-03, 5.18511662e-01], [6.71983011e-03, 6.33051968e-05, 1.89209223e-03, 5.39758130e-01], [6.24850971e-03, 5.49643175e-05, 1.88472892e-03, 6.20010927e-01]])
xg_v07 = np.array([[5.71036763e-03, 5.21517036e-05, 1.30594882e-03, 5.82846991e-01], [5.64000642e-03, 4.91813349e-05, 1.30593430e-03, 6.59152084e-01], [5.56166827e-03, 4.86554146e-05, 1.29459314e-03, 6.11729200e-01], [5.42114936e-03, 4.64831897e-05, 1.30220650e-03, 6.90426263e-01], [5.06359525e-03, 4.79161894e-05, 1.29890143e-03, 7.58683869e-01], [5.15688311e-03, 4.78794554e-05, 1.31251962e-03, 7.50973398e-01], [5.41794029e-03, 4.76190446e-05, 1.29948273e-03, 6.55835386e-01], [5.60433818e-03, 5.03199673e-05, 1.30351784e-03, 6.18930917e-01], [5.54749680e-03, 4.59438001e-05, 1.30785553e-03, 6.82751331e-01], [2.01542602e-02, 3.12873066e-03, 2.54833244e-03, 8.20638221e-01], [6.21709091e-03, 5.82590155e-05, 1.31538087e-03, 5.33207980e-01], [5.71634027e-03, 5.00150356e-05, 1.30883251e-03, 6.63619568e-01], [5.75067564e-03, 5.48429719e-05, 1.31903511e-03, 6.43545623e-01], [5.64931655e-03, 5.04590677e-05, 1.29664312e-03, 6.00031546e-01], [6.07302643e-03, 6.44282201e-05, 1.31846400e-03, 5.56895850e-01], [6.04282126e-03, 5.40277493e-05, 1.31562583e-03, 5.51196885e-01], [5.64148645e-03, 4.80319298e-05, 1.30376411e-03, 6.09373798e-01], [5.90699162e-03, 6.00830872e-05, 1.32194901e-03, 6.22856553e-01], [6.81108204e-03, 8.28886597e-05, 1.34171409e-03, 2.66799210e-01], [5.38819360e-03, 4.61480243e-05, 1.29985346e-03, 6.98788218e-01]])
procat_v07= np.array([[4.46053602e-03, 3.58894336e-05, 1.11737962e-03, 7.97252377e-01], [4.41688604e-03, 3.49843532e-05, 1.11600298e-03, 8.01757281e-01], [4.47470714e-03, 3.56920446e-05, 1.11701079e-03, 7.97408982e-01], [4.56241546e-03, 3.63625984e-05, 1.11912820e-03, 7.89192647e-01], [4.47463283e-03, 3.48130956e-05, 1.11405355e-03, 7.98465422e-01], [4.44428003e-03, 3.52207693e-05, 1.11459658e-03, 7.99225817e-01], [4.45949163e-03, 3.58503754e-05, 1.11536013e-03, 7.97475011e-01], [4.43319886e-03, 3.45965322e-05, 1.11485695e-03, 8.01728919e-01], [4.44561474e-03, 3.49250274e-05, 1.12106860e-03, 8.00139828e-01], [4.43152384e-03, 3.51038805e-05, 1.11238909e-03, 8.00811498e-01], [4.45756902e-03, 3.50201267e-05, 1.11835206e-03, 7.98102564e-01], [4.40115635e-03, 3.52164670e-05, 1.11285419e-03, 8.02673719e-01], [4.39973881e-03, 3.44768369e-05, 1.11043382e-03, 8.03247122e-01], [4.46630245e-03, 3.56421273e-05, 1.11538367e-03, 7.98424488e-01], [4.44700367e-03, 3.48456666e-05, 1.11094798e-03, 7.98876858e-01], [4.44500891e-03, 3.51381098e-05, 1.11893297e-03, 7.99885235e-01], [1.80301278e-02, 2.51374514e-03, 2.25564234e-03, 8.45385504e-01], [4.46313453e-03, 3.50883745e-05, 1.11821707e-03, 7.99260542e-01], [4.41463020e-03, 3.48281385e-05, 1.11744057e-03, 8.02691865e-01], [4.36369814e-03, 3.42412245e-05, 1.11230286e-03, 8.06201571e-01]])

# =============================================================================
# ANALYSIS LOGIC
# =============================================================================

METRICS = ["RMSE", "MSE", "MAE", "R2"]
HIGHER_IS_BETTER = {"R2"}

def to_loss(v: np.ndarray, metric: str) -> np.ndarray:
    """Convert to loss so smaller is always better for ranking."""
    return -v if metric in HIGHER_IS_BETTER else v

def p_str(p):
    """Formats p-value for display."""
    if p == 0:
        return "p < 1e-300"
    if p < 1e-3:
        return f"p = {p:.1e}"
    return f"p = {p:.3f}"

def run_friedman(entry_title: str, model_arrays: dict):
    """Runs Friedman test and returns a dictionary with results for plotting."""
    print(f"\n================ {entry_title} (Friedman only) ================\n")
    model_names = list(model_arrays.keys())
    n = next(iter(model_arrays.values())).shape[0]

    out = {"meta": {"title": entry_title}, "friedman": {}}

    for m_idx, metric in enumerate(METRICS):
        # Build loss-mapped table for ranking
        data_loss = {
            name: to_loss(model_arrays[name][:, m_idx].astype(float), metric)
            for name in model_names
        }
        df_loss = pd.DataFrame(data_loss)

        # Friedman test
        samples = [df_loss[col].values for col in df_loss.columns]
        chi2, p = friedmanchisquare(*samples)
        df = len(model_names) - 1
        out["friedman"][metric] = {"chi2": float(chi2), "df": df, "p": float(p)}

        # Calculate ranks and original means
        ranks = df_loss.rank(axis=1, method="average", ascending=True)
        avg_ranks = ranks.mean()
        means_orig = {name: float(np.mean(model_arrays[name][:, m_idx])) for name in model_names}

        summary = (
            pd.DataFrame({
                "Model": model_names,
                "Mean": [means_orig[nm] for nm in model_names],
                "AvgRank": [avg_ranks[nm] for nm in model_names],
            })
            .sort_values("AvgRank")
            .reset_index(drop=True)
        )
        out[metric] = summary

        # Print summary
        print(f"--- {metric} ---")
        print(f"Friedman χ²({df}) = {chi2:.4f}, {p_str(p)}")
        print(summary.round(6).to_string(index=False))
        print("-" * 70)
    return out

def plot_entrypoint_ranks(results_dict, metrics=("RMSE", "MSE", "MAE", "R2")):
    """
    Creates a single figure with a 1x4 panel of lollipop plots for one EntryPoint version.
    """
    title = results_dict["meta"]["title"]
    fig, axes = plt.subplots(1, len(metrics), figsize=(14, 5), constrained_layout=True)
    
    # Use a consistent color palette
    colors = plt.cm.viridis(np.linspace(0, 1, 6))

    for col, metric in enumerate(metrics):
        ax = axes[col]
        df = results_dict[metric].copy().sort_values("AvgRank")
        y = np.arange(len(df))
        
        # Create a color map based on rank
        rank_colors = [colors[int(rank-1)] for rank in df['AvgRank'].rank(method='dense')-1]

        ax.hlines(y, xmin=1, xmax=df["AvgRank"], color='gray', alpha=0.5)
        ax.plot(df["AvgRank"], y, "o", ms=8, c='black') # Markers
        # Add colored markers on top
        for i, rank_val in enumerate(df["AvgRank"]):
             ax.plot(rank_val, y[i], "o", ms=7, color=rank_colors[i])


        # Y-axis labels only on the first plot
        if col == 0:
            ax.set_yticks(y)
            ax.set_yticklabels(df["Model"], fontdict={'fontsize': 10})
        else:
            ax.set_yticks([])

        ax.invert_yaxis()
        ax.set_xlim(0.8, max(6.2, df["AvgRank"].max() + 0.2))
        ax.set_xlabel("Average Rank (lower is better)", fontsize=10)

        # Add metric title and Friedman stats
        fr = results_dict["friedman"][metric]
        ax.set_title(f"{metric}\nχ²({fr['df']})={fr['chi2']:.2f}, {p_str(fr['p'])}",
                      loc="left", fontsize=10)
        ax.grid(axis="x", alpha=0.4, linestyle="--")

    fig.suptitle(f"Model Performance Ranks for {title}", fontsize=14, fontweight="bold")
    plt.savefig(f"lollipop_rank_plot_{title.replace(' ', '_')}.png", dpi=300, bbox_inches='tight')
    plt.show()

# =============================================================================
# PACK & RUN
# =============================================================================
v06_data = {
    "OT-CatBoost": procat_v06,
    "CatBoost":     cat_v06,
    "RandomForest": random_v06,
    "LightGBM":     lgbm_v06,
    "GradBoost":    gra_v06,
    "XGBoost":      xgbo_v06,
}

v07_data = {
    "OT-CatBoost": procat_v07,
    "CatBoost":     cat_v07,
    "RandomForest": ran_v07,
    "LightGBM":     light_v07,
    "GradBoost":    grad_v07,
    "XGBoost":      xg_v07,
}

# Run analysis and get dictionaries for plotting
v06_results = run_friedman("ENTRYPOINT v06", v06_data)
v07_results = run_friedman("ENTRYPOINT v07", v07_data)

# Create two separate plots
plot_entrypoint_ranks(v06_results)
plot_entrypoint_ranks(v07_results)
```
